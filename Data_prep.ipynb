```python
import pandas as pd

students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}

scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

students = pd.DataFrame(students_data)
enrollments = pd.DataFrame(enrollments_data)
scores = pd.DataFrame(scores_data)
```

```python
print("Original Students DataFrame:")
print(students)
```

```python
print("\nNull Value Analysis:")

for col in students.columns:
    null_count = students[col].isnull().sum()
    percentage = (null_count / len(students)) * 100
    print(f"Column: {col}, Nulls: {null_count} ({percentage:.2f}%)")
```

```python
students['city'] = students['city'].fillna('Unknown')
students_clean = students.dropna(subset=['name'])
```

```python
print("\nCleaned Students DataFrame:")
print(students_clean)
```

```python
inner_join = pd.merge(students_clean, enrollments, on='student_id', how='inner')

print("\nInner Join Result:")
print(inner_join)
print("Total Rows:", len(inner_join))
```

```python
left_join = pd.merge(students_clean, enrollments, on='student_id', how='left')

print("\nLeft Join Result:")
print(left_join)
print("Total Rows:", len(left_join))
```

```python
print("\nStudents with no course:")
print(left_join[left_join['course_name'].isnull()])
```

```python
right_join = pd.merge(students_clean, enrollments, on='student_id', how='right')

print("\nRight Join Result:")
print(right_join)
print("Total Rows:", len(right_join))
```

```python
print("\nStudent IDs without names:")
print(right_join[right_join['name'].isnull()])
```

```python
outer_join = pd.merge(students_clean, enrollments, on='student_id', how='outer')

print("\nOuter Join Result:")
print(outer_join)
print("Total Rows:", len(outer_join))
```

```python
missing_rows = outer_join[(outer_join['name'].isnull()) | (outer_join['course_name'].isnull())]

print("\nRows with missing name or course:")
print(missing_rows)
```

```python
outer_indicator = pd.merge(students_clean, enrollments, on='student_id', how='outer', indicator=True)

print("\nMerge Source Distribution:")
print(outer_indicator['_merge'].value_counts())
```

```python
score_dict = dict(zip(scores['student_id'], scores['exam_score']))
```

```python
students_clean.loc[:, 'exam_score'] = students_clean['student_id'].map(score_dict)

print("\nStudents with Scores:")
print(students_clean[['student_id','name','exam_score']])
```

```python
students_no_score = students_clean.drop(columns=['exam_score'])

merge_scores = pd.merge(students_no_score, scores, on='student_id', how='left')

print("\nMerge Method Result:")
print(merge_scores[['student_id','name','exam_score']])
```

```python
def auto_merge(df1, df2, join_type, key_column):

    merged_df = pd.merge(df1, df2, how=join_type, on=key_column)

    result = {
        'result_df': merged_df,
        'row_count': len(merged_df),
        'join_type': join_type
    }

    return result
```

```python
test1 = auto_merge(students_clean, enrollments, 'inner', 'student_id')

print("Join Type:", test1['join_type'])
print("Rows in Result:", test1['row_count'])
print(test1['result_df'].head())
```

```python
test2 = auto_merge(students_clean, enrollments, 'left', 'student_id')

print("Join Type:", test2['join_type'])
print("Rows in Result:", test2['row_count'])
print(test2['result_df'].head())
```
